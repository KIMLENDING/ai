```python
%matplotlib inline
fig = plt.figure(figsize=(6,6))
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)

for i in range(64):
    ax = fig.add_subplot(8, 8, i + 1, xticks=[],yticks=[])
    ax.imshow(digits.images[i],cmap=plt.cm.binary, interpolation='nearest')
    ax.text(0, 7, str(digits.target[i]))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target)
```

```python
from sklearn.naive_bayes import GaussianNB
OneOfNavieBayesianModels = GaussianNB()
```

```python
OneOfNavieBayesianModels.fit(X_train, y_train)
```

```python
predicted = OneOfNavieBayesianModels.predict(X_test)
```

```python
#Plot the prediotion
fig = plt.figure(figsize=(6,6))
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)

#plot the digits: ech image is 8x8 pocels
for i in range(64):
    ax = fig.add_subplot(8, 8, i + 1, xticks=[],yticks=[])
    ax.imshow(X_test.reshape(-1,8,8)[i], cmap=plt.cm.binary, interpolation='nearest')
    
    if predicted[i]== y_test[i]:
        ax.text(0, 7, str(digits.target[i]), color='green')
    else:
         ax.text(0, 7, str(digits.target[i]), color='red')
```
